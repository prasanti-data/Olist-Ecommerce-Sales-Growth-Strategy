MERGING DATASETS

```python
#joining products and product category
product_merged = pd.merge(products,product_category, on = 'product_category_name', how = 'inner')
```

```python
product_merged.head()
```

```python
product_merged.info()
```

```python
product_merged['Product_Volume'] = product_merged['product_length_cm'] * product_merged['product_height_cm'] * product_merged['product_width_cm']
```

```python
product_merged = product_merged.drop(columns = ['product_length_cm','product_height_cm','product_width_cm'])
```

```python
product_merged = product_merged.drop(columns = ['product_category_name'])
```

```python
product_merged.info()
```

MERGING ORDER_ITEM AND PRODUCT_MERGED TABLE ON PRODUCT ID

```python
Order_product_merged = pd.merge(order_items,product_merged, on = 'product_id', how = 'inner')
```

```python
Order_product_merged.info()
```

MERGING ORDER_PRODUCT_MERGED WITH ORDERS TABLE ON ORDER ID


```python
Order_Product_merged_1 = pd.merge(Order_product_merged,order, on = 'order_id', how = 'inner')
```

```python
Order_Product_merged_1.info()
```

MERGING ORDER_PRODUCT_MERGED_1 WITH ORDER PAYMENTS TABLE ON ORDER ID


```python

Order_Product_merged_2 = pd.merge(Order_Product_merged_1,order_payments, on = 'order_id', how = 'inner')
```

```python
Order_Product_merged_2.info()
```

```python
Order_Product_merged_3 = pd.merge(Order_Product_merged_2,order_reviews, on = 'order_id', how = 'left')
```

```python
Order_Product_merged_3.info()
```

JOINING ORDER MERGED DATA WITH CUSTOMERS DATA

```python
Order_Product_Customer = pd.merge(Order_Product_merged_3,customers, on = 'customer_id', how = 'inner')
```

```python
Order_Product_Customer.info()
```

JOINING PRODUCT_CUSTOMER MERGED DATA WITH SELLER DATA

```python
Final_df1 = pd.merge(Order_Product_Customer,sellers, on = 'seller_id', how = 'inner')
```
